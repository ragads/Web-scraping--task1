```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

urls=["https://www.nature.com/articles/s41598-025-91196-5",
      "https://www.nature.com/articles/s41598-024-83651-6",
      "https://www.nature.com/articles/s41598-025-90010-6",
      "https://www.nature.com/articles/s41598-025-89937-7",
      "https://www.nature.com/articles/s41598-025-86860-9",
      "https://www.nature.com/articles/s41598-025-89777-5",
      "https://www.nature.com/articles/s41598-025-90321-8",
      "https://www.nature.com/articles/s41598-024-67426-7",
      "https://www.nature.com/articles/s41586-024-07609-4",
      "https://www.nature.com/articles/s41598-025-85448-7",
      "https://www.nature.com/articles/s41598-024-63544-4",
      "https://www.nature.com/articles/s41578-023-00636-z",
      "https://www.nature.com/articles/s41529-025-00553-2",
      "https://www.nature.com/articles/s41598-025-85912-4",
      "https://www.nature.com/articles/s41598-025-86288-1",
      "https://www.nature.com/articles/s41467-024-49361-3",
      "https://www.nature.com/articles/s41598-025-86271-w",
      "https://www.nature.com/articles/s43246-025-00739-w",
      "https://www.nature.com/articles/s41598-025-86446-5",
      "https://www.nature.com/articles/s41467-024-49210-3",
      "https://www.nature.com/articles/s41598-025-85990-4",
      "https://www.nature.com/articles/s41598-024-84453-6",
      "https://www.nature.com/articles/s41467-024-47954-6",
      "https://www.nature.com/articles/s41598-024-85086-5",
      "https://www.nature.com/articles/s41467-023-44331-7",
      "https://www.nature.com/articles/s41598-024-81753-9",
      "https://www.nature.com/articles/s41467-024-46967-5",
      "https://www.nature.com/articles/s41598-024-83809-2",
      "https://www.nature.com/articles/s41467-024-46591-3",
      "https://www.nature.com/articles/s41598-024-82020-7",
      "https://www.nature.com/articles/s41598-024-83857-8",
      "https://www.nature.com/articles/s41598-023-48954-0",
      "https://www.nature.com/articles/s41598-024-83812-7",
      "https://www.nature.com/articles/s41598-024-83103-1",
      "https://www.nature.com/articles/s41598-023-43646-1",
      "https://www.nature.com/articles/s41598-023-45810-z",
      "https://www.nature.com/articles/s41598-023-46483-4",
      "https://www.nature.com/articles/s41467-023-40381-z",
      "https://www.nature.com/articles/s41598-024-80540-w",
      "https://www.nature.com/articles/s41598-024-83243-4",
      "https://www.nature.com/articles/s41598-024-83088-x",
      "https://www.nature.com/articles/s41598-024-81808-x",
      "https://www.nature.com/articles/s41598-024-83176-y",
      "https://www.nature.com/articles/s41598-024-80785-5",
      "https://www.nature.com/articles/s41467-023-42213-6",
      "https://www.nature.com/articles/s41467-022-33936-z",
      "https://www.nature.com/articles/s44172-024-00329-7",
      "https://www.nature.com/articles/s41586-023-06419-4",
      "https://www.nature.com/articles/s41467-023-40752-6",
      "https://www.nature.com/articles/s41598-025-90321-8"

]


HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}




def scrape_data(url):
  try:
    response = requests.get(url, headers=HEADERS)
    response.raise_for_status()
    soup=BeautifulSoup(response.content,"html.parser")
    news_title=soup.find("h1",attrs={"class":"c-article-title"}).text.strip()
    news_description=soup.find("div",attrs={"class":"c-article-section__content"}).text.strip()
    authors=soup.find_all("a",{"data-test": "author-name"})
    author_names = [author.text.strip() for author in authors]
    date=soup.find("time",attrs={"datetime":True}).text.strip()
    webiste_name="Nature"

    return{
        "News Title":news_title,
        "News Description":news_description,
        "Author Name":author_names,
       "Date":date,
        "Website Name":webiste_name,
        "url":url
    }
  except Exception as e:
    print(f"Error scraping data from {url}: {e}")
    return None

news=[]
for url in urls:
  data=scrape_data(url)
  if data:
    news.append(data)

data_df=pd.DataFrame(news)
data_df.to_csv("Aerospace.csv",index=False)

data_df
```
